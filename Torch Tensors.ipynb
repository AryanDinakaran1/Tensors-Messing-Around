```python
# BARE BONES CALCULATIONS
test_tensor_1 = torch.tensor([1,2,3,4,5])
test_tensor_1 += 10
test_tensor_1 *= 10
test_tensor_1 -= 10


test_tensor_1 * 5
```

```python
# BUILT-IN CALCULATION FUNCTIONS
test_tensor_2 = torch.tensor([1,2,3,4,5], device='cpu')

torch.add(test_tensor_2, 5)
torch.sub(test_tensor_2, 5)
torch.mul(test_tensor_2, 5)
torch.div(test_tensor_2, 5)
```

```python
# ELEMENT WISE MULTIPLICATION
tensor_1_zz = torch.tensor([1,2,3,4,5])
tensor_2_zz = torch.tensor([5,4,3,2,1])

tensor_1_zz * tensor_2_zz
```

```python
# MATRIX MULTIPLICATION
MATRIX_1 = torch.rand([16,5], dtype=torch.float32, device="cpu")
MATRIX_2 = torch.rand([5,16], dtype=torch.float32, device="cpu")

RAND_IMAGE = torch.mm(MATRIX_1, MATRIX_2)
RAND_IMAGE.shape
```

```python
MATRIX_1 = torch.rand([2,3], dtype=torch.float32, device="cpu")
MATRIX_2 = torch.rand([2,3], dtype=torch.float32, device="cpu")

torch.mm(MATRIX_1, MATRIX_2.T)
```
